## FIFA WORLD CUP ANALYSIS

```python
# Importing the dependencies
import pandas as pd
import numpy as np
```

```python
# Loading the Dataset
worldcup_matches = pd.read_csv('/content/WorldCupMatches.csv')
worldcup_players = pd.read_csv('/content/WorldCupPlayers.csv')
worldcup_info = pd.read_csv('/content/WorldCups.csv')
```

```python
worldcup_matches.head()
```

```python
worldcup_players.head()
```

```python
worldcup_info.head()
```

```python
# Data Integration: Merging Dataset
merged_data = pd.merge(worldcup_matches, worldcup_players, on = ['RoundID', 'MatchID'])
merged_data = pd.merge(merged_data, worldcup_info, on = 'Year')
```

```python
# Checking the info the merged dataset
merged_data.info()
```

```python
# Checking missing value
merged_data.isnull().sum()
```

```python
# Dropping irrelevant columns and null values
merged_data.dropna(inplace = True)
```

```python
merged_data.isnull().sum()
```

```python
merged_data.describe()
```

```python
# Exporting merged data to CSV
merged_data.to_csv('merged_data.csv', index = False)
```
